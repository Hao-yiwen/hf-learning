```python
from transformers import pipeline

# 创建文本生成pipeline，使用DeepSeek-R1-Distill-Qwen-1.5B模型
pipe = pipeline(
    "text-generation",
    model="deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B",
    trust_remote_code=True,
    device_map="auto"
)

# 测试生成文本
prompt = "讲个睡前故事"
result = pipe(
    prompt,
    max_new_tokens=4096,
    temperature=0.7,
    do_sample=True,
    pad_token_id=pipe.tokenizer.eos_token_id
)

print(result[0]['generated_text'])
```
